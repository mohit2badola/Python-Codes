Making matrix with the help of list

```python
mat=[]
lm=int(input("Enter the length of column"))
lr=int(input("Enter the length of rows"))
for i in range(lm):
    mat.append([])
    for j in range(lr):
        val=int(input("Enter the element"))
        mat[i].append(val)

print(mat)
```

filtering a nested list or mstrix

adding all the odd elements of the matrix in another list

```python
m = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
odd=[]

even=[]
for i in m:
    for j in i:
        if j%2!=0:
            odd.append(j)

        else:
            even.append(j)
print("The odd list is ",odd)
print("The even list is ",even)
```

making a nested list into a single falt list

```python
m = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

nl=[]

for i in m:
    for j in i:
        nl.append(j)
print(nl)
```

## Remove duplicates from a list without using set().

```python
ls=[1,22,2,1,4,144,45,424,324,2,2,2,2,1232,4234,1,24,2132,4234,131]
nls=[]
for i in ls:
    if i not in nls:
        nls.append(i)

print("the list after removing the nested elements is", nls)
```

## Merge two lists and sort without using sort().

```python
# l1=[1,23,4,5,6,34,424,42,242,43,354,2]
l1=[]
l2=[41,233,24,25,62,324,424,242,2422,243,354,22]

l=l1+l2

print(f"The list after merging both the list is {l}")

c=len(l)
for i in range(c):
    for j in range(0,c-i-1):
        if l[j]>l[j+1]:
            l[j],l[j+1]=l[j+1],l[j]
print(l)
```

## Find all pairs whose sum equals a target value. 

```python
l=[1,2,3,4,5,6,7,8,6,9,4,2]

target=7
ln=[]
for i in range(len(l)):
    for j in range(i,len(l)):
        if l[i]+l[j]==target :
            if (l[i],l[j]) not in ln:
                ln.append((l[i],l[j]))
print(ln)

type(ln[0])
```

## Find difference between largest and smallest element.

```python
l=[41,233,24,25,62,324,424,242,2422,243,354,22]
sml=l[0]
large=l[0]

for i in range(len(l)):
    if l[i]<sml:
        sml=l[i]
    elif l[i]>large:
        large=l[i]

print(f"The difference between the smallest and largest elemment is {large-sml}")
```
