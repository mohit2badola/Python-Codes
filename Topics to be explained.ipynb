### What is a List?
A lsit is a collection of data types stored as a single value

a list is :
 - ordered
 - Mutable
 - hetrogenious



There is no fixed size for a list basically

An empty list size in a 32 bit arcitecture is 36 bytes

for a list with datamembers it is usually the 36 byte plus the size of each varible as their pointer is stored 
in general we can say that:

    size of list in memory-->  ** 32+(4*n)   : where n is the number of elements**

```python
l=[1,2,3,4,5,6]
l1=[1,'abc',True,32.5]
a="Mohit"
l2=list(a)
print(l)
print(l1)
print(l2)
print(list())
```
