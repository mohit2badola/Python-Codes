Iterating in a dictionary 

```python
data=dict(a="Alpha",b="beta",c="gamma")
```

```python
for key in data:
    print(key)
```

```python
for value in data.values():
    print(value)
```

```python
for key,value in data.items():
    print(f"{key},{value}")
```

c. key: Print marks of a given key in a statement as "Marks of student name is : marks".

```python
students=['Rohit','Vinay','Tarun','Subham']
marks=[60,80,50,90]
```
